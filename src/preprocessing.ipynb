```python
import numpy as np
import pandas as pd
import datetime
```

```python
df = pd.read_csv('../datasets/nasa.csv')

# Dropped because they are identifiers, duplicate data or single value columns:
dropped_columns = [
    'Neo Reference ID',
    'Name',
    'Est Dia in KM(min)',
    'Est Dia in KM(max)',
    'Est Dia in Miles(min)',
    'Est Dia in Miles(max)',
    'Est Dia in Feet(min)',
    'Est Dia in Feet(max)',
    'Close Approach Date',
    'Relative Velocity km per hr',
    'Miles per hour',
    'Miss Dist.(miles)',
    'Miss Dist.(kilometers)',
    'Miss Dist.(Astronomical)',
    'Orbiting Body',
    'Orbit ID',
    'Equinox',
]

# Potentially unnecessary columns:
unnecessary_columns = [
    'Absolute Magnitude',
    'Orbit Determination Date',
]

df.drop(dropped_columns, axis=1, inplace=True)
df.drop(unnecessary_columns, axis=1, inplace=True)

# Data preprocessing continues:
df['Epoch Date Close Approach'] = df['Epoch Date Close Approach'] / 1000
# df['Orbit Determination Date'] = pd.to_datetime(df['Orbit Determination Date']).astype('int64') / 10**9

print(df.columns)
print(df.head())

# Save to CSV:
df.to_csv('../datasets/nasa_cleaned.csv', index=False)
```
